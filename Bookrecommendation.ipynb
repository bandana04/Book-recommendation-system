```python
ratings=pd.read_csv('/content/drive/MyDrive/BX-Book-Ratings.csv',sep=";",on_bad_lines='skip',encoding='latin-1')
```

```python
x=ratings['user_id'].value_counts()>50
```

```python
y=x[x].index
```

```python
y
```

```python
ratings=ratings[ratings['user_id'].isin(y)]
```

```python
ratings.shape
```

```python
ratings.head()
```

```python
ratings_with_books=ratings.merge(books, on='ISBN')
```

```python
ratings_with_books.shape
```

```python
number_rating=ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.rename(columns={'rating':'number of ratings'},inplace=True)
```

```python
number_rating.head()
```

```python
final_rating=ratings_with_books.merge(number_rating,on='title')
```

```python
final_rating
```

```python
final_rating=final_rating[final_rating['number of ratings']>=10]
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final_rating.shape
```

```python
book_pivot=final_rating.pivot_table(columns='user_id',index='title',values='rating')
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse=csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distances,suggestions=model.kneighbors(book_pivot.iloc[352,:].values.reshape(1,-1), n_neighbors=6)
```

```python
suggestions
```

```python
for i in range(len(suggestions)):
  print(book_pivot.index[suggestions[i]])
```

```python
book_pivot.index[352]
```

```python
np.where(book_pivot.index=='I Is for Innocent')[0][0]
```

```python

def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distances, suggestions = model.kneighbors(book_pivot.iloc[book_id, :].values.reshape(1, -1), n_neighbors=6)

    for i in range(len(suggestions)):
        print(book_pivot.index[suggestions[i]])
```

```python
recommend_book('I Is for Innocent')
```
